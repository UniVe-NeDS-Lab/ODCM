```python
import glob
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
```

```python
dists = []
for f in glob.glob('../data/*'):
    g = nx.read_graphml(f)
    for s,t in g.edges():
        dists.append(g[s][t]['dist'])
dists = np.array(dists)
```

```python
len(dists[dists<2000])
```

```python
len(dists[dists<2000])/len(dists)
```

```python
plt.hist(dists, bins=range(0,12000,1000))
```
